```python
####
from selenium import webdriver
from selenium.webdriver.chrome.options import Options  ## for headless mode
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.support.select import Select
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
#######
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
import re 
from matplotlib import pyplot as plt
import seaborn as sns
import urllib3
pd.plotting.register_matplotlib_converters()
from warnings import filterwarnings
filterwarnings('ignore')
# %matplotlib inline
import time
print("Setup Complete")
```

### function

```python
def scraper(url_of_page):
    num=[]
    df=pd.DataFrame()
    http = urllib3.PoolManager()
    l = http.request('GET',url_of_page)
    soup_1=BeautifulSoup(l.data, 'html.parser')
    links_1 = soup_1.findAll('a', class_='phone')
    for i in range (len(links_1)):
        number=re.sub('<[^>]+>' , ' ', str(links_1)).strip().replace(' (','').replace('[','').replace(']','').replace(') ','').replace('-','').split(',')[0].strip()
        num.append({'link':url_of_page,'number':number})
    df=pd.DataFrame(num)
    # df.to_excel('output1.xlsx', engine='xlsxwriter')
    return df
```

```python
scraper('https://turbo.az/autos/6735686-hyundai-sonata')
```

## personal auto

```python
%%time
links_personal_auto=[]
num=[]
df=pd.DataFrame()
for x in range(1,2):
    url=f'https://turbo.az/autos?page={x}'
    request=requests.get(url).content
    soup=BeautifulSoup(request,'html.parser')
    cars=soup.find_all('div',class_='products-i')
    for i in cars:
        links_personal_auto.append('https://turbo.az'+i.a['href'])      
    for i in range (len(links_personal_auto)):
        http = urllib3.PoolManager()
        content = http.request('GET',links_personal_auto[i])
        soup_1=BeautifulSoup(content.data, 'html.parser')
        try:
            links_1 = soup_1.findAll('div',class_="seller-name")
            
            name=re.sub('<[^>]+>' , ' ', str(links_1)).strip().replace(' (','').replace('[ ','').replace(' ]','').replace(') ','').replace('-','').split(',')[0].strip()
            
        except:
            
            name='null'
       
        try:
            
            links_2 = soup_1.findAll('a', class_='phone')
            
            number=int(re.sub('<[^>]+>' , ' ', str(links_2)).strip().replace(' (','').replace('[','').replace(']','').replace(') ','').replace('-','').split(',')[0].strip())
            
        except :
        
            number=0
        try:
            
            links_3 = soup_1.findAll('h1', class_='product-name product-name-row')
            
            automobile=re.sub('<[^>]+>' , ' ', str(links_3)).strip().replace(' (','').replace('[ ','').replace(' ]','').replace(') ','').replace('-','').split(',')[0].strip()
            
        except:
            
            automobile='null'
        try:
            
            links_4 = soup_1.findAll('div', class_='product-price')
            
            price=re.sub('<[^>]+>' , ' ', str(links_4)).strip().replace(' (','').replace('[ ','').replace(' ]','').replace(') ','').replace('-','').split(',')[0].strip()
            
        except:
            
            price='null'
            
        try:
            
            links_5 = soup_1.findAll('div', class_='product-properties-value')
            
            city=re.sub('<[^>]+>' , ' ', str(links_5)).strip().replace(' (','').replace('[ ','').replace(' ]','').replace(') ','').replace('-','').split(',')[0]
            
        except:
            
            city='null'
            
        num.append({'url':links_personal_auto[i],'name':name,'number':number,'automobile':automobile,'price':price,'city':city})

    df=pd.DataFrame(num)        
df 
```

```python
df.to_excel('personal_numbers.xlsx', engine='xlsxwriter')
```
